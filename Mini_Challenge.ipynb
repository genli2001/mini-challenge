```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import pandas as pd
from PIL import Image
import os
import numpy as np
import cv2
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
```

```python
#load y data, using the corresponding numeric value 0 - 100 from the category.csv for manipulation
categories = pd.read_csv('category.csv')
train = pd.read_csv('train.csv')

train['Category'] = train['Category'].str.strip().str.lower()
categories.iloc[:, 1] = categories.iloc[:, 1].str.strip().str.lower()

# create a mapping from names to numbers
category_mapping = dict(zip(categories.iloc[:, 1], categories.iloc[:, 0]))

# Apply the mapping to assign numbers to names in 'train_small.csv'
train['category_number'] = train['Category'].map(category_mapping)
train['category_number'] = train['category_number'].astype('int8')

y = train['category_number']
#np.save('ymain_array.npy', y)
```

```python
##The code in this section about face recognition is taken from: https://towardsdatascience.com/face-detection-in-2-minutes-using-opencv-python-90f89d7c0f81

#Note: due to time constraint for the deadline, and the memory constraint I have, only 66842 data successfully loaded, 
# and due to the memory constraint for training, only 11 * 3518 images were used for training, which is why you see
# data are split into chunks
csv_path = 'train.csv'
df = pd.read_csv(csv_path)
filenames = df['File Name'].values

path = 'train'
X = []
a = 0

face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# create the X: load images, convert to grayscale, detect face, then crop, normalize, and resize
for filename in filenames:
    
    img_path = os.path.join(path, filename)

    loaded_image = Image.open(img_path).convert('RGB')
    img_array = np.array(loaded_image)  # Original RGB image
    
    img_gray = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)
    faces = face_cascade.detectMultiScale(img_gray, 1.1, 4)
    
    # crop from the RGB image, keep original if no face detected
    if len(faces) > 0:
        x, y, w, h = faces[0]
        cropped_face = img_array[y:y+h, x:x+w]
    else:
        cropped_face = img_array

    resized = cv2.resize(cropped_face, (100, 100))
    
    # normalize 
    normalized_face = (resized / 255.0).astype(np.float32)
    X.append(normalized_face)
    a += 1
    print('Processed', a)

X = np.array(X)
np.save('Xmain_array.npy', X)
```

```python
test_path = 'test' 
X_test = []
b = 0
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')


filenames = [f for f in os.listdir(test_path) if f.endswith('.jpg')]
filenames.sort(key=lambda f: int(f.split('.')[0]))  # Sort filenames based on the numerical part

for filename in filenames:
    img_path = os.path.join(test_path, filename)
    
    loaded_image = Image.open(img_path).convert('RGB')
    img_array = np.array(loaded_image) 
    
    img_gray = cv2.cvtColor(img_array, cv2.COLOR_RGB2GRAY)
    faces = face_cascade.detectMultiScale(img_gray, 1.1, 4)
    
    if len(faces) > 0:
        x, y, w, h = faces[0]
        cropped_face = img_array[y:y+h, x:x+w]  
    else:
        cropped_face = img_array 

    resized = cv2.resize(cropped_face, (100, 100))
    normalized_face = (resized / 255.0).astype(np.float32)
    
    X_test.append(normalized_face)
    b +=1
    print('process', b)

X_test = np.array(X_test)
np.save('Xtest_array.npy', X_test)
```

```python
#The CNN model
model = models.Sequential([

    layers.experimental.preprocessing.RandomContrast(0.15),  
    layers.experimental.preprocessing.RandomRotation(factor=0.2),  
    layers.experimental.preprocessing.RandomFlip("horizontal"),
    
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(100, 100, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),
    layers.Dropout(0.2),

    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),

    layers.Conv2D(256, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.BatchNormalization(),

    layers.GlobalAveragePooling2D(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.4), 
    layers.Dense(100, activation='softmax') 
])

model.compile(optimizer= Adam(learning_rate=0.001),
                loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
```
